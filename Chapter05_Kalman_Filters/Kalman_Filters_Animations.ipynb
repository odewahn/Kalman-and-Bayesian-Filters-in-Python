```python
from matplotlib import animation
import matplotlib.pyplot as plt

def animate(filename, func, frames, interval, figsize=(6.5, 6.5)):
    def forward(frame):
        # I don't know if it is a bug or what, but FuncAnimation calls twice
        # with the first frame number. That breaks any animation that uses
        # the frame number in computations
        if forward.first:
            forward.first = False
            return
        func(frame)
    
    fig = plt.figure(figsize=figsize)
    forward.first = True
    anim = animation.FuncAnimation(fig, forward, frames=frames, interval=interval)
    anim.save(filename, writer='imagemagick')
```

```python
from __future__ import print_function, division

import numpy.random as random
import math
import stats

def update(mu1, var1, mu2, var2):
    mean = (var1*mu2 + var2*mu1) / (var1+var2)
    variance = 1 / (1/var1 + 1/var2)
    return (mean, variance)

def predict(pos, variance, movement, movement_variance):
    return (pos + movement, variance + movement_variance)


sensor_error = 1.2**2
movement_error = .2
movement = 0
voltage = (25,20) #who knows what the first value is?

volts = [14.44, 18.74, 17.21, 15.98, 16.76, 14.8, 16.01, 16.90, 15.41, 17.31,
        16.39, 17.96]
ps = [voltage[0]]


i = 0
pred = 0
def volt_animate(frame):
    global i, ps, voltage, Z, x, pred
    
    step = frame % 4
    plt.subplot(211)
    plt.ylim([14,26])

    if step == 0:
        prev = voltage[0]
        voltage = predict(voltage[0], voltage[1], movement, movement_error)
        pred = voltage[0]
        plt.plot ([i, i+1], [prev, pred], c='g') 
        
    elif step == 1:
        Z = volts[i]
        i += 1
        
        plt.scatter(i, Z, marker='+', s=64, color='r')

    elif step == 2:
        plt.plot ([i,i], [pred, Z], c='r', alpha=0.3)

    else:
        voltage = update(voltage[0], voltage[1], Z, sensor_error)
        ps.append(voltage[0])
        plt.plot(ps, c='b')
        
    plt.subplot(212)
    plt.cla()
    stats.plot_gaussian(voltage[0], voltage[1], xlim=[0,32])
    plt.ylim([0,1])
    plt.tight_layout()
        
N=12
animate('volt_animate.gif', volt_animate, N*3, 500)
```

```python
class DogSensor(object):
    
    def __init__(self, x0=0, velocity=1, noise=0.0):
        """ x0 - initial position
            velocity - (+=right, -=left)
            noise - scaling factor for noise, 0== no noise
        """
        self.x = x0
        self.velocity = velocity
        self.noise = math.sqrt(noise)

    def sense(self):
        self.x = self.x + self.velocity
        return self.x + random.randn() * self.noise
    
    
# assume dog is always moving 1m to the right
movement = 1
movement_error = .2
sensor_error = 4.5
pos = (0, 100)   # gaussian N(0,50)

dog = DogSensor(pos[0], velocity=movement, noise=sensor_error)

zs = []
ps = []
vs = []

def dog_animate(frame):
    global pos, zs, ps, vs
    pos = predict(pos[0], pos[1], movement, movement_error)    
    Z = dog.sense()
    zs.append(Z)
    vs.append(pos[1])
    print(pos[1])
    
    pos = update(pos[0], pos[1], Z, sensor_error)
    ps.append(pos[0])

    plt.subplot(211)
    plt.scatter (frame, Z)
    plt.subplot(212)
    plt.cla()
    stats.plot_gaussian(pos[0], pos[1], xlim=[0,32])
    plt.ylim([0,1])


N=12
animate('dog_track.gif', dog_animate, N, 200)
```
