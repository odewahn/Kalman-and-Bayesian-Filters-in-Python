```python
from matplotlib import animation
import matplotlib.pyplot as plt

def animate(filename, func, frames, interval, figsize=(6.5, 6.5)):
    def forward(frame):
        # I don't know if it is a bug or what, but FuncAnimation calls twice
        # with the first frame number. That breaks any animation that uses
        # the frame number in computations
        if forward.first:
            forward.first = False
            return
        func(frame)
    
    fig = plt.figure(figsize=figsize)
    forward.first = True
    anim = animation.FuncAnimation(fig, forward, frames=frames, interval=interval)
    anim.save(filename, writer='imagemagick')
```

```python
from __future__ import print_function, division

import numpy.random as random
import math

def update(mu1, var1, mu2, var2):
    mean = (var1*mu2 + var2*mu1) / (var1+var2)
    variance = 1 / (1/var1 + 1/var2)
    return (mean, variance)

def predict(pos, variance, movement, movement_variance):
    return (pos + movement, variance + movement_variance)


sensor_error = 1.2**2
movement_error = 0
movement = 0
voltage = (19,20) #who knows what the first value is?

volts = [14.44, 18.74, 17.21, 15.98, 16.76, 16.25, 16.01, 16.90, 15.41, 17.31,
        16.39, 17.96]
ps = [voltage[0]]


i = 0
pred = 0
def volt_animate(frame):
    global i, ps, voltage, Z, x, pred
    
    step = frame % 4
    plt.ylim([12,20])

    if step == 1:
        Z = volts[i]
        i += 1
        
        plt.scatter(i, Z, marker='+', s=64, color='r')

    elif step == 3:
        voltage = update(voltage[0], voltage[1], Z, sensor_error)
        ps.append(voltage[0])
        plt.plot(ps, c='b')
        
    elif step == 2:
        plt.plot ([i,i], [pred, Z], c='r', alpha=0.3)

    else:
        prev = voltage[0]
        voltage = predict(voltage[0], voltage[1], movement, movement_error)
        pred = voltage[0]
        plt.plot ([i, i+1], [prev, pred], c='g') 
        
N=12       
animate('volt_animate.gif', volt_animate, N*3, 600)
```
